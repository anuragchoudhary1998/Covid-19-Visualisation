```python
data.to_csv('modified.csv')
data.to_csv('modified.csv')
india = dt[dt['location'] == 'India']
usa = dt[dt['location'] == 'United States']
uk = dt[dt['location'] == 'United Kingdom']
```

```python
india['total_deaths']=india['total_deaths'].fillna(0)
usa['total_deaths']=usa['total_deaths'].fillna(0)
uk['total_deaths']=uk['total_deaths'].fillna(0)
```

```python
india['new_deaths']=india['new_deaths'].fillna(0)
usa['new_deaths']=usa['new_deaths'].fillna(0)
uk['new_deaths']=uk['new_deaths'].fillna(0)
```

```python
india['people_vaccinated']=india['people_vaccinated'].fillna(0)
usa['people_vaccinated']=usa['people_vaccinated'].fillna(0)
uk['people_vaccinated']=uk['people_vaccinated'].fillna(0)
```

```python
india['new_cases']=india['new_cases'].fillna(0)
usa['new_cases']=usa['new_cases'].fillna(0)
uk['new_cases']=uk['new_cases'].fillna(0)
```

```python
india.isnull().sum()
usa.isnull().sum()
uk.isnull().sum()
```

```python
india['date'] = pd.to_datetime(india['date'], utc=True)
usa['date'] = pd.to_datetime(usa['date'], utc=True)
uk['date'] = pd.to_datetime(uk['date'], utc=True)
india = india.set_index('date')
usa = usa.set_index('date')
uk = uk.set_index('date')
```
